```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import ggplot, aes, geom_density, geom_line, geom_point, ggtitle
from sklearn.neighbors import LocalOutlierFactor #Outliar handling
import webcolors
from colormap import rgb2hex
import ast
from tqdm import tqdm, tqdm_notebook
tqdm_notebook().pandas()
```

```python
#read csv file
df = pd.read_csv("data/images_colors_updated.csv")
df.head()
```

```python
# a function that returns the name of the closest color name to the (requested_color)
# requested_color is in RGB format

def closest_color(requested_color):
    min_colours = {}
    for key, name in webcolors.CSS3_HEX_TO_NAMES.items():
        r_c, g_c, b_c = webcolors.hex_to_rgb(key)
        rd = (r_c - requested_color[0]) ** 2
        gd = (g_c - requested_color[1]) ** 2
        bd = (b_c - requested_color[2]) ** 2
        min_colours[(rd + gd + bd)] = name
    return min_colours[min(min_colours.keys())]
```

```python
# a function that returns the name of the (requested_color) & calls closest color function if the color is not found

def get_color_name(requested_colour):
    try:
        closest_name = actual_name = webcolors.rgb_to_name(requested_colour)
    except ValueError:
        closest_name = closest_color(requested_colour)
        actual_name = None
    if (actual_name == None):
        return closest_name
    else:
        return actual_name
```

```python
# changing the tyoe of dcolor to a tuple 
df['dcolor'] = df['dcolor'].apply(ast.literal_eval)
```

```python
# calling the get color name function on the column
df['dcname'] = df['dcolor'].progress_apply(get_color_name)
```

```python
#saving to csv file
df.to_csv('full_emotion_color.csv', index = False)
```
